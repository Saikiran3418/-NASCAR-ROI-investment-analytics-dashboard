```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
df.isnull().sum()
```

```python
import numpy as np

np.random.seed(42)

# Sponsor Cost based on performance
df["Sponsor_Cost"] = np.where(
    df["AvgFinish"] <= 10,
    np.random.randint(150000, 250001, len(df)),
    np.where(
        df["AvgFinish"] <= 20,
        np.random.randint(90000, 150001, len(df)),
        np.random.randint(50000, 90001, len(df))
    )
)
```

```python
# Social Impressions (Exposure)
df["Social_Impressions"] = (
    df["Wins"] * 50000 +
    df["LapsLed"] * 20 +
    df["Points"] * 30 +
    np.random.randint(50000, 200000, len(df))
).astype(int)

# Brand Mentions (Engagement)
df["Brand_Mentions"] = (
    df["Social_Impressions"] * np.random.uniform(0.01, 0.05, len(df))
).astype(int)
```

```python
# Value creation
df["Media_Value"] = df["Social_Impressions"] * 0.08
df["Engagement_Value"] = df["Brand_Mentions"] * 2.5

# ROI
df["ROI"] = (
    (df["Media_Value"] + df["Engagement_Value"] - df["Sponsor_Cost"])
    / df["Sponsor_Cost"]
)
```

```python
import numpy as np

# Reuse core business metrics
df["Exposure"] = df["Social_Impressions"]
df["Engagement_Rate"] = df["Brand_Mentions"] / df["Social_Impressions"]
df["Performance_Score"] = 40 - df["AvgFinish"]

# Safe normalization function
def minmax(col):
    if col.max() == col.min():
        return pd.Series([0.5] * len(col), index=col.index)
    return (col - col.min()) / (col.max() - col.min())

# Normalize metrics
df["ROI_norm"] = minmax(df["ROI"])
df["Exposure_norm"] = minmax(df["Exposure"])
df["Engagement_norm"] = minmax(df["Engagement_Rate"])
df["Performance_norm"] = minmax(df["Performance_Score"])

# Sponsor Target Score
df["Sponsor_Target_Score"] = (
    df["ROI_norm"] * 0.35 +
    df["Exposure_norm"] * 0.25 +
    df["Engagement_norm"] * 0.20 +
    df["Performance_norm"] * 0.20
)

# Rank drivers
df["Target_Rank"] = df["Sponsor_Target_Score"].rank(ascending=False, method="dense")

# Create sponsor target category
df["Sponsor_Target_Category"] = np.where(
    df["Sponsor_Target_Score"] >= 0.70, "High Priority Target",
    np.where(
        df["Sponsor_Target_Score"] >= 0.45, "Medium Priority Target",
        "Low Priority Target"
    )
)

# Show best sponsorship targets
top_targets = df[[
    "Driver", "Year", "AvgFinish", "Wins", "Points",
    "ROI", "Sponsor_Target_Score", "Target_Rank", "Sponsor_Target_Category"
]].sort_values(by="Sponsor_Target_Score", ascending=False)

print(top_targets.head(10))
```

```python
import pandas as pd
import numpy as np

# Safe normalization function
def minmax(col):
    if col.max() == col.min():
        return pd.Series([0.5] * len(col), index=col.index)
    return (col - col.min()) / (col.max() - col.min())

# -----------------------------
# RETURN METRIC
# -----------------------------
df["Return_Score"] = minmax(df["ROI"])

# -----------------------------
# RISK COMPONENTS
# -----------------------------
# Higher AvgFinish is worse, so it increases risk
df["Finish_Risk"] = minmax(df["AvgFinish"])

# Lower wins means more risk, so invert wins
df["Wins_Risk"] = 1 - minmax(df["Wins"])

# Lower points means more risk
df["Points_Risk"] = 1 - minmax(df["Points"])

# Higher sponsor cost means more financial risk
df["Cost_Risk"] = minmax(df["Sponsor_Cost"])

# -----------------------------
# FINAL RISK SCORE
# -----------------------------
df["Risk_Score"] = (
    df["Finish_Risk"] * 0.35 +
    df["Wins_Risk"] * 0.25 +
    df["Points_Risk"] * 0.20 +
    df["Cost_Risk"] * 0.20
)

# -----------------------------
# RISK CATEGORY
# -----------------------------
df["Risk_Category"] = np.where(
    df["Risk_Score"] >= 0.70, "High Risk",
    np.where(df["Risk_Score"] >= 0.40, "Medium Risk", "Low Risk")
)

# -----------------------------
# RETURN CATEGORY
# -----------------------------
df["Return_Category"] = np.where(
    df["Return_Score"] >= 0.70, "High Return",
    np.where(df["Return_Score"] >= 0.40, "Medium Return", "Low Return")
)

# -----------------------------
# RISK vs RETURN SEGMENT
# -----------------------------
def risk_return_segment(row):
    if row["Risk_Score"] < 0.40 and row["Return_Score"] >= 0.70:
        return "Best Bet"
    elif row["Risk_Score"] >= 0.40 and row["Return_Score"] >= 0.70:
        return "Aggressive Opportunity"
    elif row["Risk_Score"] < 0.40 and row["Return_Score"] < 0.70:
        return "Safe but Moderate"
    else:
        return "High Risk / Low Return"

df["Risk_Return_Segment"] = df.apply(risk_return_segment, axis=1)

# -----------------------------
# FINAL VIEW
# -----------------------------
risk_return_view = df[[
    "Driver", "Year", "AvgFinish", "Wins", "Points",
    "Sponsor_Cost", "ROI", "Risk_Score", "Return_Score",
    "Risk_Category", "Return_Category", "Risk_Return_Segment"
]].sort_values(by=["Return_Score", "Risk_Score"], ascending=[False, True])

print(risk_return_view.head(15))
```
